## Performing segmentation on the given data

```python
from sklearn.preprocessing import LabelEncoder
number = LabelEncoder()
df_new["Subsidy for charging station establishment"] = number.fit_transform(df_new["Subsidy for charging station establishment"].astype("str"))
df_new["EV buying incentives"] = number.fit_transform(df_new["EV buying incentives"].astype("str"))
df_new["EV distribution target in policy "] = number.fit_transform(df_new["EV distribution target in policy "].astype("str"))
df_new["Power Plant target in policy"] = number.fit_transform(df_new["Power Plant target in policy"].astype("str"))
```

```python
df_new
```

## One hot encoding the categorical data

```python
data_onehot = df_new.copy()
data_onehot = pd.get_dummies(data_onehot,columns = ["Subsidy for charging station establishment","EV buying incentives","EV distribution target in policy ","Power Plant target in policy"])
```

```python
data_onehot
```

```python
data_onehot = data_onehot.drop(columns = data_onehot.columns[0])
```

## Scaling the above data

```python
from sklearn.preprocessing import StandardScaler
```

```python
SS = StandardScaler()
data_scaled = SS.fit_transform(data_onehot)
```

```python
data_scaled
```

## Reducing the dimensionality of data

```python
from sklearn.decomposition import PCA
```

```python
import numpy as np
```

```python
pca = PCA().fit(data_scaled)
plt.figure(figsize = (10,10))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.title("Explained Variance")
plt.xlabel("Number of components")
plt.ylabel("Percentage of Each component")
```

```python
pca = PCA(n_components = 3)
pca.fit(data_scaled)
data_pca = pca.transform(data_scaled)
```

```python
data_pca
```

## K means clustering by considering all varoables as segmentation variables

```python
import warnings 
warnings.filterwarnings("ignore")
from sklearn.cluster import KMeans
Sum_of_squared_distance = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters = k)
    km = km.fit(data_pca)
    Sum_of_squared_distance.append(km.inertia_)
plt.plot(K,Sum_of_squared_distance,"bx-")
plt.xlabel("K")
plt.ylabel("Sum of Squared distance")
plt.title("Elbow method for optimal K")
```

```python
#elbow formation optimal = 4
kmeans = KMeans(n_clusters = 4,init = "k-means++",n_init = 10,max_iter = 1000,tol = 0.0001,verbose = 0,random_state = None,copy_x = True,
               algorithm = "auto")
y_means = kmeans.fit_predict(data_pca)
```

```python
y_means
```

## Attaching this cluster information with the data

```python
df_new["cluster"] = y_means
```

## Clustering by fitting GMM on PCA data

```python
from sklearn.mixture import GaussianMixture
```

```python
n_components = np.arange(1,15)
models = [GaussianMixture(n,covariance_type = "full",random_state = 0).fit(data_pca) for n in n_components]
plt.figure(figsize = (10,8))
plt.plot(n_components,[m.bic(data_pca) for m in models],label = 'BIC')
plt.plot(n_components,[m.aic(data_pca) for m in models],label = 'AIC')
plt.legend(loc = "best")
plt.xlabel("Number of components")
```

## We can see that BIC predicts around 5 components of the data

```python
gmm = GaussianMixture(n_components = 5)
gmm.fit(data_pca)
```

```python
y_predict = gmm.predict(data_pca)
```

## Importance of Segmentation:
###  This segmentation in important as it helps us classify states on the basis of their average enrgy consumption per year and the number of charging stations established by the PSUs on behalf of the state government. It gives us an indirect estimate of the number of electric vehicles running on the roads of a given state till now. Moreover, this segmentation analysis also covers the qualitative aspects of the policies designed by various state governments in order to ensure effecient adoption of eletric vehicles by their public. It also helps us recognize the states(geographical regions) which are really interested in adoption of EV vehicles by their public apart from the others as in their policies, some of the states have set a target of setting certain minimum amount of charging stations by 2025 and they also ensure in their plolicy that certain minimum fixed percentage of the state registered vehicles will be electric vehicle. Including these factors in the dataset gives us an idea of awareness among states for adoption of electric vehicles. This would in turn enable any EV startup to target proper market segment to establish their base.

## CONCLUSION

### From this segmentation analysis we find that most of the north eastern states along with Jammu and Kashmir, Chhattisgarg and Jharkhand are unsuitable for establishment of an EV startup due to lack of awareness of governments of the corresponding states regarding the adoption of Electric vehicles in their region. It is better to switch to central states like Delhi,Rajasthan, Bihar or states of the south like Maharashtra,Telangana and Karnataka as their government seems to be more aware and involved in process of effective adoption of EVs by their public.These staes are also charcetrized by high consumption of power and higher charging stations(established by PSUs) and would therefore be a suitable choice for production of EVs.This result was visible after analysing all the segmentation techniques on the given data.
